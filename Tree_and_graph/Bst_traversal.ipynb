```python
from collections import deque
class node:
    def __init__(self,data):
        self.data=data
        self.left=None
        self.right=None

class Bst:
    def __init__(self):
        self.root=None

    def add(self,num):
        nn=node(num)
        if self.root==None:
            self.root=nn
            return
        trav=self.root

        while True:
            if num<trav.data:
                if trav.left is None:
                    trav.left=nn
                    break
                trav=trav.left
            else:
                if trav.right is None:
                    trav.right=nn
                    break
                trav=trav.right
    
    def inorder(self,proc):
        if proc:
            self.inorder(proc.left)
            print(proc.data,end=" ")
            self.inorder(proc.right)

    def preorder(self,proc):
        if proc:
            print(proc.data,end=" ")
            self.preorder(proc.left)
            self.preorder(proc.right)
    def postorder(self,proc):
        if proc:
            self.postorder(proc.left)
            self.postorder(proc.right)
            print(proc.data,end=" ")

    def levelorder(self,proc):
        if proc is None:
            return []
        ans=[]
        queue=deque([proc])

        while queue:
            curr=[]
            for i in range(len(queue)):
                a=queue.popleft()
                curr.append(a.data)
                if a.left:
                    queue.append(a.left)
                if a.right:
                    queue.append(a.right)
            ans.append(curr)
        return ans
    
    #iterative methods
    #preorder
    def iterative_preorder(self,proc):
        if proc is None:
            return []
        ans=[]
        stack=[proc]
        while stack:
            a=stack.pop()
            ans.append(a.data)
            if a.right:
                stack.append(a.right)
            if a.left:
                stack.append(a.left)
        return ans

sol=Bst()
sol.add(10)
sol.add(5)
sol.add(1)
sol.add(6)
sol.add(20)
sol.add(30)
sol.add(15)


#sol.inorder(sol.root)
#sol.preorder(sol.root)
#sol.postorder(sol.root)
#sol.levelorder(sol.root)
sol.iterative_preorder(sol.root)
```
